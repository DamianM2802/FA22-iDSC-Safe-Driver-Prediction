## The Relationship and Difference of 5 Types of Variables

First of all, in order to protect customers' information, this project does not disclose the specific meaning of these 58 analysis features of the data. In other words, we cannot perform an intuitive analysis from the [column name], and we cannot tell which features are effective and useful for prediction!

For example: I want to predict whether a person is male or female. We may decide by many intuitive dimensions such as hair length, height, and weight. We will evaluate the degree of influence (weight) of the characteristics according to natural consciousness and experience. Usually such approach turns out to be effective.

However, the meanings of all dimensions here are pretty vague, and we cannot make subjective judgments about them. We can only disassemble and integrate them through different variable types, looking through them to finding the hidden connections.

With `df.info()`, we can tell there are 2 data types for distinct variables. -`int64` and  `float64`, what does this mean?

Intuitively, this means that the data has been proccesed before releasing, and the all the data has been converted to integers (also to protect customers privacy).

Then, for these 58 features, what are the possible data tyes for them?

**5 Major Data Types (also known as: Levels of Measurement)**

Briefly asÔºö

- Binary ==> 0 or 1
    - only 2 outputsÔºö0/1

- Nominal ==> categorical, non-ordered
    - most common data typeÔºå**not ordered**
    - e.g. a list of country names (China, US, UK, Canada etc.) **there is no specific ordering between them.**

- Ordinal ==> categorical, but ordered
    - just like orderd nominal
    - e.g. grade ranking of students (Emma 1stÔºåLily 2ndÔºåJohn 3rd etc.) **the ordering here means the ranking of thier grades, ordinal**

- Interval ==> continuous, do not have an absolute zero point
    - countinuous dataÔºåcan be included in stat calculations to compare magnitude and quantify measurements
    - e.g. temperature (we can say 40C > 20CÔºåbut we cannot say it is twice 20C) **interval data does not have an absolute zero pointÔºÅ**
    
- Ratio ==> continuous, acommadate a zero point
    - countinuous data, can be included in any stat calculations to compare magnitude and quantify measurements
    - e.g. height (200cm > 100cm, and we can also say 200cm = 2*100cm) **ratio data has an absolute zero pointÔºÅ**

### Metadata

You may not be familiar with this concept, but since it is such a heated topic for recent years, you may have heard about it!


- What is MetadataÔºü

MetadataÔºå also known as the data about data itself.

- Why it matters in many projectsÔºü

When working on projects, we often encounter data with quite high dimensions and many features. In order to do further analysis on them, we must organize these data in a structured way, and Metadata could be a good start for this.

- So, what does it actaully doÔºü

It mainly describes the properties of data (a analysis of the organization and structure of data), also of great significance for future data analysis, visualization, and modeling.
